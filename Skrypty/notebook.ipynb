```python
from sympy import *
from robotyka import *
q1, q2, q3, q4 = symbols('q1 q2 q3 q4', real=True)  # q
a, a1, a2, a3 = symbols('a a1 a2 a3', real=True)    # a
d, d1, d2, d3 = symbols('d d1 d2 d3', real=True)    # d
alfa, beta = symbols('\\alpha \\beta', real=True)
x, y, z = symbols('x y z')                          # osie
#
# rot(z,q1)*trans(z,a1)*trans(x,a)*rot(z,pi/2)
```

```python
A01 = rot(z,q1)*tr(z,d)*rot(x,pi/2)
A01
```

```python
A12 = rot(z,alfa)*rot(x,pi/2)
A12
```

```python
A23 = tr(z,q3)
A23
```

```python
A03 = A01*A12*A23 
A03 # A03
```

```python
K_kart = A03.col(3)
K_kart.row_del(3)
K_kart # Kinematyka dla katezjańskichfsefsefsefsetqedfsfsefsefse
```
