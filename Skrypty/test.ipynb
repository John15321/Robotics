```python
import numpy as numpy
import matplotlib.pyplot as plt
from sympy import *
from sympy import print_latex
from rotations import *
```

```python
alpha, beta, gamma, theta = sp.symbols('\\alpha \\beta \\gamma \\theta', real=True)
q1, q2, q3, l1, l2, t, a1, a2, a3, d = sp.symbols('q1 q2 q3 l1 l2 t a1 a2 a3 d', real=True)
z, x, y = sp.symbols(' z x y', real=True)
```

```python
A01 = rot(z, q1)
A01
```

```python
A12 = trans(z, a2)*rot(x, sp.pi/2)
A12
```

```python
A23 = trans(z, q3)
A23
```

```python
A03 = A01*A12*A23
A03
```

```python
K_kart = A03.col(3)
K_kart.row_del(3)
K_kart # Kinematyka dla katezjańskich
```

```python
Jak = Matrix()
Jak = Jak.col_insert(0,Derivative(K_kart,q1).doit())
Jak = Jak.col_insert(2,Derivative(K_kart,q2).doit())
Jak = Jak.col_insert(2,Derivative(K_kart,q3).doit())
Jak # Jakobian ylko dla kartezjańskich
```
